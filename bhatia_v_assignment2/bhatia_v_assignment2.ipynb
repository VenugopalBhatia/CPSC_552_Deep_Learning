```python
metric_array,trained_model = f.train(error_type="CE")
```

The Cross Entropy Loss function gives a higher test accuracy of 76.8% while the mean square error gives a test accuracy of 70.65%. It also seems convergence is faster in the Cross Entropy loss function since we can see that by epoch 400 the test accuracy reaches 73.7% after which by epoch 900 we get an improvement of about 2%. In Comparison MSE makes a jump of 2% from epoch 400 to epoch 500, and we see an improvement of about 2% after every 100 epochs. The rationale here could be that the MSE might be non convex here resulting in a local minima being found.
